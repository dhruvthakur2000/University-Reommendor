```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
def extract(page):
#     headers = {'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'}
    url = f'https://www.thegradcafe.com/survey/index.php?page={page}&per_page=9'
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    cards = soup.find_all('div', class_ = 'row mb-2')
    for card in cards:
        
        htag = card.h6
        htag.select('span', class_ = "text-secondary ms-3" )[0]
        name = htag.contents[0].strip()
#         print(name)
        
        info_card = card.find('div', class_ = 'mt-3')
        span_card = info_card.find('span').text.strip()
        span_card = re.sub(r'\n\t*',"",span_card)
#         print(span_card)
    
        x = ''
        for item in info_card.find_all('span', 'badge badge-unselected'):
            try:
                info = item.text.strip()
                info = info.ljust(len(info)+1)
#                 print(info)
                x+=info

#                 print(item.text.strip())
            except:
                pass
#         print(info_card.text)
        data = {
            'Name':name,
            'Admit_Status':span_card,
            'Info':x
        }
        
        con_list.append(data)
    
        
con_list = []

for page in range(48001,50001):
    c= extract(page)
    print(f"getting page...{page}")
#     print(con_list)

```

```python
def clean_dataframe(df):
    # Cleaning Info Col
#     df = df[~df['Info'].str.contains("GRE AW")==False]
#     df = df[df['Info'].str.contains("PhD")==False]

    # Splitting University and Major
    df[['Major','College', 'School','Univ_Addition', 'Univ1',
        'Univ2', 'Univ3', 'Univ4', 'Univ5','Univ6']]=df['Name'].str.split(pat = ',',expand = True)
#         ,
#        'Univ7', 'Univ8','Univ9', 'Univ10',
#         'Univ11', 'Univ12', 'Univ13', 'Univ14', 'Univ15','Univ16',
#        'Univ17'
    df.head()

    # Splitting Admit Status
    df[['Status', 'On', 'Date','Month']] = df['Admit_Status'].str.split(' ', expand = True)
    df.drop(columns = ['On','Date','Month'], axis = 1,inplace = True)
    # df.head()

    # Splitting Info Column
    df[['Semester', 'Year','Citizenship', 'GRE_Q', 'Q_Score','GRE_V','V','V_Score',
        'GRE_AWA','AW', 'AWA_Score','GPA',
        'GPA_Score','Program', '0','1']] = df['Info'].str.split(' ', expand = True)
    df.drop(columns = ['GRE_Q','GRE_V','V','GRE_AWA','AW','GPA','0','1'], axis = 1,inplace = True)
    # df.head()

    # Cleaning New Columns
    df = df[df['Q_Score'].str.contains("GRE")==False]
    df = df[df['Q_Score'].str.contains("AW")==False]
    df = df[df['Q_Score'].str.contains("V")==False]

    df = df[df['AWA_Score'].str.contains("Masters")==False]
    df = df[df['Program'].str.contains("MFA")==False]
    df = df[df['Program'].str.contains("Other")==False]
    df = df[df['Program'].str.contains("PsyD")==False]

    # Dropping Original Cols
    df.drop(columns = ['Name', 'Admit_Status', 'Info'], inplace = True)
    # df1.head()

    # Filtering Q_Score
    df['Q_Score'] = df['Q_Score'].astype(int)
    df['V_Score'] = df['V_Score'].astype(int)

    df.drop(df.loc[(df['Q_Score']>170)].index, inplace = True)
    df.drop(df.loc[(df['Q_Score']<130)].index, inplace = True)
    df.drop(df.loc[(df['V_Score']<130)].index, inplace = True)
    df.drop(df.loc[(df['V_Score']>170)].index, inplace = True)
    df.drop(df.loc[(df['AWA_Score']>'6.0')].index, inplace = True)

    df['Program'] = df['Program'].replace("", "Masters")
    df['School'] = df['School'].fillna("")
    df['Univ_Addition'] = df['Univ_Addition'].fillna("")
    df['Univ1'] = df['Univ1'].fillna("")
    df['Univ2'] = df['Univ2'].fillna("")
    df['Univ3'] = df['Univ3'].fillna("")
    df['Univ4'] = df['Univ4'].fillna("")
    df['Univ5'] = df['Univ5'].fillna("")
    df['Univ6'] = df['Univ6'].fillna("")
#     df['Univ7'] = df['Univ7'].fillna("")
#     df['Univ8'] = df['Univ8'].fillna("")
#     df['Univ9'] = df['Univ9'].fillna("")
#     df['Univ10'] = df['Univ10'].fillna("")
#     df['Univ11'] = df['Univ11'].fillna("")
#     df['Univ12'] = df['Univ12'].fillna("")
#     df['Univ13'] = df['Univ13'].fillna("")
#     df['Univ14'] = df['Univ14'].fillna("")
#     df['Univ15'] = df['Univ15'].fillna("")
#     df['Univ16'] = df['Univ16'].fillna("")
#     df['Univ17'] = df['Univ17'].fillna("")
    
    
    df['University'] = df['College']+df['School']+df['Univ_Addition']+df['Univ1']
    +df['Univ2'] + df['Univ3'] + df['Univ4'] + df['Univ5'] + df['Univ6']
#      + df['Univ7']
#     +df['Univ8'] + df['Univ9'] + df['Univ10'] + df['Univ11'] + df['Univ12']
#     + df['Univ13']+ df['Univ14']+ df['Univ15'] + df['Univ16'] + df['Univ17']

    df.drop(columns = ['College','School','Univ_Addition',
                       'Univ1','Univ2', 'Univ3','Univ4', 'Univ5','Univ6'], axis = 1,inplace = True)

#                        , 'Univ4',
#                       'Univ5', 'Univ6', 'Univ7', 'Univ8','Univ9',
#                        'Univ10', 'Univ11', 'Univ12',
#                       'Univ13', 'Univ14', 'Univ15', 'Univ16', 'Univ17'
                      
    df = df[['University', 'Major', 'Program', 'Citizenship',
              'Semester', 'Year', 'Q_Score', 'V_Score', 'AWA_Score', 'GPA_Score','Status']]

    df.reset_index(drop=True, inplace = True)

    df['GPA_Score'] = df['GPA_Score'].replace('','0.0')
    df['GPA_Score'] = df['GPA_Score'].astype(float)
    df['GPA_Score'] = df['GPA_Score'].round(1)
    
    df.drop_duplicates(keep=False, inplace = True)

    return df
```
